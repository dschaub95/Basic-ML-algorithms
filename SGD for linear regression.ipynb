```python
import numpy as np

x = np.arange(0,1,0.01)
epsilon = (np.random.rand(1,100)-0.5)*0.6
y = (np.sin(2*np.pi*x)+epsilon)[0]
print(epsilon)
```

```python
def model_poly(theta_, x):
    result = 0
    for i in range(0,theta_.shape[0]):
        result += theta_[i]*np.power(x, i)
    return result
```

```python
def error(val1, val2):
    return np.sqrt(np.sum((1/x.shape[0])*np.square(val1-val2)))
    
```

```python

    

theta = np.random.rand(1,4)[0]-0.5

alpha = 0.12
err_list = []
for k in range(0,10000):
    for i in range(0, x.shape[0]):
        for j in range(0,theta.shape[0]):
            theta[j] += alpha*(y[i]-model_poly(theta, x[i]))*(x[i]**j)
    err = error(y, model_poly(theta, x))
    err_list.append(err)
```

```python
import matplotlib.pyplot as plt

plt.plot(err_list)
plt.xlabel('Iterations')
plt.ylabel('Root Mean Square Error')
plt.ylim(0,1)

plt.savefig('Error', format='pdf')
```

```python
plt.plot(x, y, 'bo', markersize=3, label='Data')
plt.plot(x, model_poly(theta, x), label='Model')
plt.plot(x, np.sin(np.pi*2*x), label='Sinus')
plt.xlabel('X-Value')
plt.ylabel('Y-Value')
plt.legend()
plt.savefig('Graph', format='pdf')
```
