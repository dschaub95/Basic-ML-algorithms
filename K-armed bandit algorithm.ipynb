```python
import random
import matplotlib.pyplot as plt
import numpy as np


# predefine the mean of the rewards each action on the bandit has --> this defines the optimal action
Q_star = np.zeros((10,))
for a in range(0,10):
    Q_star[a] = np.random.normal(0,1)

run = 0

#average_list = [None, None, None]
#percent_list = [None, None, None]
#optimal_list = [None, None, None]

for epsilon in [0]:
    
    # initialize average reward and percent optimal action
    average_reward = np.zeros((1000,))
    percent_optimal_actions = np.zeros((1000,))
    num_optimal_actions = np.zeros((1000,))
    
    # averaging of 2000 runs
    for i in range(0,2000):
        
        # predefine the mean of the rewards each action on the bandit has --> this defines the optimal action
        Q_star = np.zeros((10,))
        for a in range(0,10):
            Q_star[a] = np.random.normal(0,1)

        
        # approximate values for the 
        Q = np.zeros((10,))
        N = np.zeros((10,))

        # optimal action counter
        count = 0

        # number of plays
        for t in range(0,1000):
            # 0 with probability epsilon
            s = np.random.binomial(1,epsilon)

            if s == 1:
                # greedy action
                for k in range(0,9):
                    action = np.argmax(Q)
            else:
                # select random action
                action = random.randint(0,9)



            # update stuff
            R = np.random.normal(Q_star[action],1)
            N[action] += 1 
            Q[action] += (1/N[action])*(R-Q[action])
            
            # count number of times the optimal action was selected
            if action == np.argmax(Q_star):
                count += 1
            percent_optimal_actions[t] += (1/2000)*(count/np.sum(N))
            num_optimal_actions[t]+= (1/2000)*(count)   
            
            # update average reward (check again)
            tmp = 0
            for idx, q in enumerate(Q):
                tmp += q*N[idx]
            average_reward[t] += (1/2000)*(tmp/np.sum(N))
            
            #average_reward[t] += (1/2000)*np.max(Q)
            
            
    # update big lists for each epsilon
    average_list[0] = average_reward
    percent_list[0] = percent_optimal_actions
    optimal_list[0] = num_optimal_actions
    
    
    
    
    
    run += 1
    print(run)
```
